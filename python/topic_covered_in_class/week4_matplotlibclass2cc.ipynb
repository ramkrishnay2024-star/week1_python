**Advance Histogram**

```python
import matplotlib.pyplot as plt
import numpy as np

blood_sample_men = [112, 90, 76, 94, 50, 70, 113, 140, 150, 160, 139, 120, 76, 77, 99, 88, 110, 67, 50, 80]
blood_sample_women = [132, 80, 66, 64, 90, 80, 143, 150, 130, 140, 119, 60, 76, 97, 69, 128, 150, 167, 150, 180]
blood_sample_kids = [132, 80, 66, 64, 90, 80, 143, 150, 130, 140, 119, 60, 76, 97, 69, 128, 150, 167, 150, 180]

plt.hist([blood_sample_men, blood_sample_women, blood_sample_kids], bins = [50,75,100,125,150,175,200], color = ['green','orange','blue'], label = ['men', 'women', 'kids'])
plt.legend()
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.xticks([50,75,100,125,150,175,200])
plt.yticks(np.arange(1,9))
plt.title("Blood Sugar Chart")
plt.show()
```

```python
plt.hist([blood_sample_men, blood_sample_women], bins = [75,100,125,150], color = ['green', 'orange'], label = ['men', 'women'], orientation = 'horizontal')
plt.legend()
plt.ylabel("Sugar Level")
plt.xlabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.show()
```

```python
# Create data

x = ["A", "B", "C", "D"]
y1 = np.array([10, 20, 10, 30])
print(sum(y1))
y2 = np.array([20, 25, 15, 25])
print(sum(y2))
y3 = np.array([12, 15, 19, 6])
print(sum(y3))
y4 = np.array([10, 29, 13, 19])
print(sum(y4))

# plot bars in stack manner
plt.bar(x, y1, color = 'r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='g')
plt.bar(x, y4, bottom=y1+y2+y3, color='c')

plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.show()
```

```python
# Create data

x = ["A", "B", "C", "D"]
y1 = np.array([10, 20, 10, 30])
print(sum(y1))
y2 = np.array([20, 25, 15, 25])
print(sum(y2))
y3 = np.array([12, 15, 19, 6])
print(sum(y3))
y4 = np.array([10, 29, 13, 19])
print(sum(y4))

# plot bars in stack manner
plt.barh(x, y1, color = 'r')
plt.barh(x, y2, left=y1, color='b')
plt.barh(x, y3, left=y1+y2, color='g')
plt.barh(x, y4, left=y1+y2+y3, color='c')

plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.show()
```

```python
import numpy as np
x = [10, 20, 30, 40, 50, 60, 70, 120]
x1 = [10, 20, 40, 50, 30, 60, 90]
print("PYTHON MEAN:", np.mean(x))
print("PYTHON MEDIAN:", np.median(x))
print("C++ MEAN:", np.mean(x1))
print("C++ MEDIAN:", np.median(x1))
y = [x, x1]

plt.boxplot(y, notch=False, orientation='vertical', widths=0.7, tick_labels=['PYTHON', "C++"], 
            showmeans=True, boxprops=dict(color='blue'), capprops=dict(color="red"), 
            sym="x", whiskerprops=dict(color="yellow"))

plt.show()
```
